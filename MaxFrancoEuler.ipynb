1. If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

    Find the sum of all the multiples of 3 or 5 below 1000.

```python
def sum_nat_numbers(length= 1000):
    my_list = []
    for n in range(length):
        if (n % 3 == 0 ) or (n % 5 == 0):
            my_list.append(n)
    return sum(my_list)

sum_nat_numbers()
```

```python
def even_fibonacci():
    #first we need to create a list with the value 1 since the sequence starts at 1
    #its used to store the sequence
    fib= [1]
    
    #Create fibonacci sequence
    while fib[-1] < 4000000:
        if fib[-1] == 1:
            fib.append(2)
        else:
            fib.append(fib[-1] + fib[-2])    
    
    #Create a an empty list to fill only the even numbers of the fib list
    my_list=[]
    
    #check for even values & then append only even values
    for i in fib:
        if i%2 == 0:
            my_list.append(i)
    #Return sum of values
    return sum(my_list)

```

6. The sum of the squares of the first ten natural numbers is,

    12 + 22 + ... + 102 = 385
    The square of the sum of the first ten natural numbers is,

    (1 + 2 + ... + 10)2 = 552 = 3025
    Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is 
    3025 − 385 = 2640.

    Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

```python
def sum_of_squares(nat_num=100):
    squaresum= sum(np.array(range(1,nat_num+1))**2)
    sumsquare= sum(range(1,nat_num+1))**2
    
    diff = sumsquare - squaresum
    return diff
    
```

9. A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,

    a2 + b2 = c2
    For example, 32 + 42 = 9 + 16 = 25 = 52.

    There exists exactly one Pythagorean triplet for which a + b + c = 1000.
    Find the product abc.

```python
def pythagorean_triplet():
    for a in range(1,1000): #range where the number a could be
        for b in range(1,1000): #range where the number b could be
            c = 1000-a-b #condition the a+b+c = 1000
            if (a**2+b**2) == c**2: #condition that its pythogorean triplet
                return a*b*c
```
